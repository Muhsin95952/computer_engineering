Frequency Table of Gender

```python
data = {"male": 100,
        "female": 60}

gender = list(data.keys())
freq = list(data.values())
explode = (0, 0.05)
color = ["lightcoral", "gold"]


plt.pie(freq, explode = explode, labels = gender, colors = color,  shadow = True, startangle = 90)
plt.axis('equal')
```
